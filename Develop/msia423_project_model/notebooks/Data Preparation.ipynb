# How Data is Prepared for Model

```python
import pandas as pd
import pickle
```

## Loading combined crime and location data

```python
ungroupedDF = pd.read_csv("../data/interim/OriginalMergedZipcode.csv")
```

```python
ungroupedDF.head()
```

```python
ungroupedDF = ungroupedDF.drop("Updated On", 1)
```

```python
len(ungroupedDF)
```

### We aren't as interested in domestic crime, which shouldn't exhibit any seasonality anyway, so we drop it.

```python
ungroupedDF = ungroupedDF[ungroupedDF["Domestic"] != True]
```

Exploring the categories of crime committed in the column "Primary Type"

```python
ungroupedDF["Primary Type"].unique()
```

```python
ungroupedDF[ungroupedDF["Primary Type"] == "OTHER OFFENSE"]#["Description"].unique()
```

```python
ungroupedDF[ungroupedDF["Primary Type"] == 'NON-CRIMINAL']["Description"].unique()
```

```python
ungroupedDF[ungroupedDF["Primary Type"] == "NON - CRIMINAL"]["Description"].unique()
```

```python
excludeList = ["DECEPTIVE PRACTICE", "PROSTITUTION", "LIQUOR LAW VIOLATION", 
              "GAMBLING", "NON-CRIMINAL", "NON - CRIMINAL", "NON-CRIMINAL (SUBJECT SPECIFIED)"]
ungroupedDF = ungroupedDF[~ungroupedDF["Primary Type"].isin(excludeList)]
```

```python
len(ungroupedDF)
```

```python
ungroupedDF["Date"] = ungroupedDF["Date"].str.slice(stop= 10)
```

```python
ungroupedDF["Date"] = pd.to_datetime(ungroupedDF["Date"], format="%m/%d/%Y")
```

```python
ungroupedDF["Month"] = ungroupedDF.Date.dt.month
```

```python
#Reading in the files
weatherDFs = [] 
for i in range(1,11):
    weatherDFs.append(pd.read_csv("../data/external/WeatherHist" + str(i)+ ".csv", header = 0))
fullWeatherDF = pd.concat(weatherDFs)
```

```python
fullWeatherDF.head()
```

```python
fullWeatherDF["Date"] = pd.to_datetime(fullWeatherDF["Date"], format="%m/%d/%Y")
```

```python
fullWeatherDF.head()
```

```python
#ungroupedDF = ungroupedDF.merge(fullWeatherDF, on="Date" ) #Left join by default
```

```python
ungroupedDF.head()
```

```python
#groupedDF = ungroupedDF.groupby(by = [pd.Grouper("Day"), "Zipcode"])["Description"].count()
groupedDF = ungroupedDF.groupby(by = ["Date", "Zipcode"])["Description"].count()
```

```python
groupedDF = groupedDF.reset_index()
```

```python
groupedDF= groupedDF.merge(fullWeatherDF, on= "Date") #pd.DataFrame(groupedDF)#.merge(fullWeatherDF, on= "Date")
```
